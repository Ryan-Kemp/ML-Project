```python
from pyboy import PyBoy

# Load the game
pyboy = PyBoy(r"C:\Users\rkemp\Downloads\Pokemon_ Red Version\Pokemon - Red Version (USA, Europe) (SGB Enhanced).gb")

# Helper: advance the emulator a number of ticks
def advance(frames=60):
    for _ in range(frames):
        pyboy.tick()

# Helper: press and release a button
def press(button, hold_frames=1, release_frames=1):
    pyboy.send_input(button)      # press
    advance(hold_frames)
    pyboy.send_input(-button)     # release
    advance(release_frames)

# Example usage
advance(1800)        # let the game boot up
press(7)            # START button (open menu/title screen)
advance(100)        # wait a bit

# Walk right 10 steps
for _ in range(10):
    press(3)        # RIGHT button
```
